```python
import pandas as pd
from pathlib import Path

PROCESSED_DIR = Path(r"D:\1 Univesrity work\Lect\Data  management\US Retail Sales and Public Holidays Analysis\DATA\processed")

orders_path = PROCESSED_DIR / "orders_clean.csv"
holidays_path = PROCESSED_DIR / "holidays_clean.csv"

orders = pd.read_csv(orders_path, parse_dates=['Order Date'])
holidays = pd.read_csv(holidays_path, parse_dates=['date'])

# Quick checking the data
print("Orders shape:", orders.shape)
print("Holidays shape:", holidays.shape)
print(orders.head())
print(holidays.head())
```

```python
import pandas as pd

# Ensure 'Order Date' and 'date' columns are datetime (just in case)
orders['Order Date'] = pd.to_datetime(orders['Order Date'])
holidays['date'] = pd.to_datetime(holidays['date'])

# Merge orders with holidays on date
merged_df = orders.merge(
    holidays[['date', 'name']], 
    how='left',
    left_on='Order Date',
    right_on='date'
)

# Drop the extra date
merged_df = merged_df.drop(columns=['date'])

# Create a boolean column: True if it's a holiday, False otherwise
merged_df['IsHoliday'] = ~merged_df['name'].isna()

merged_df = merged_df.rename(columns={'name': 'Holiday Name'})

# Quick check
print(merged_df[['Order ID', 'Order Date', 'Holiday Name', 'IsHoliday']].head(10))
```
