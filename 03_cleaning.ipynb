```python
duplicate_holidays = holidays_df.duplicated(subset=['date'])
print("Number of duplicate holiday dates:", duplicate_holidays.sum())
```

```python
holidays_df = holidays_df.drop_duplicates(subset=['date'])
print("After removing duplicates, holidays shape:", holidays_df.shape)
```
