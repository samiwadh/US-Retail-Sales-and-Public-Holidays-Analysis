```python
# Q4 Test whether adding holiday information improves daily order prediction accuracy.import pandas as pd

import pandas as pd
import numpy as np

# Ensure Order Date is datetime
merged_df['Order Date'] = pd.to_datetime(merged_df['Order Date'])

# Filter 2014-2017
merged_df = merged_df[(merged_df['Order Date'] >= '2014-01-01') &
                      (merged_df['Order Date'] <= '2017-12-31')]

# Aggregate daily orders
daily_data = merged_df.groupby('Order Date').agg({
    'IsHoliday': 'max'
}).reset_index()

daily_counts = merged_df.groupby('Order Date').size().reset_index(name='order_count')
daily_data = daily_data.merge(daily_counts, on='Order Date')

# Add features
daily_data['day_of_week'] = daily_data['Order Date'].dt.dayofweek
daily_data['month'] = daily_data['Order Date'].dt.month
daily_data['day'] = daily_data['Order Date'].dt.day

daily_data = daily_data.sort_values('Order Date').reset_index(drop=True)
```

```python
split_idx = int(len(daily_data) * 0.8)
train = daily_data.iloc[:split_idx]
test = daily_data.iloc[split_idx:]

# Features
X_train_base = train[['day_of_week', 'month', 'day']]
X_test_base = test[['day_of_week', 'month', 'day']]

X_train_holiday = train[['day_of_week', 'month', 'day', 'IsHoliday']]
X_test_holiday = test[['day_of_week', 'month', 'day', 'IsHoliday']]

y_train = train['order_count']
y_test = test['order_count']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Baseline (no holiday)
model_base = LinearRegression().fit(X_train_base, y_train)
y_pred_base = model_base.predict(X_test_base)

# With holiday
model_holiday = LinearRegression().fit(X_train_holiday, y_train)
y_pred_holiday = model_holiday.predict(X_test_holiday)

# Accuracy metrics
mae_base = mean_absolute_error(y_test, y_pred_base)
rmse_base = np.sqrt(mean_squared_error(y_test, y_pred_base))

mae_holiday = mean_absolute_error(y_test, y_pred_holiday)
rmse_holiday = np.sqrt(mean_squared_error(y_test, y_pred_holiday))

print("Baseline Model (no holiday): MAE =", round(mae_base,2), ", RMSE =", round(rmse_base,2))
print("Holiday-Enhanced Model: MAE =", round(mae_holiday,2), ", RMSE =", round(rmse_holiday,2))
```
